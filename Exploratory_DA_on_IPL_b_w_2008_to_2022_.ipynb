#Data cleaning

###Checking nulls

```
#Checking missing values
ipl.isnull().sum()
```

```
#in percentage
ipl.isnull().sum()/len(ipl)*100
```

```
#method column is more than 95 % missing values we can drop,also team 1 players and team2players
ipl.drop(columns=["method","Team1Players","Team2Players"],axis=1,inplace=True)
```

```
#Now we have four columns which has missing values so handlw accordingly
#Handling city with mode of the column
ipl["City"].mode()
```

```
#also we can see the mode by
ipl["City"].value_counts()[:5]
```

```
#Going to replace city column with mode values
ipl["City"].fillna("Mumbai", inplace = True)
```

```
#Mode value for superover
ipl["SuperOver"].mode()
```

```
#Changing super over with mode
ipl["SuperOver"].fillna("N", inplace= True)
```

```
#Mode value for winningteam
ipl["WinningTeam"].mode()
```

```
#Replacing the value
ipl["WinningTeam"].fillna("Mumbai Indians", inplace= True)
```

```
#Mode for player of match
ipl["Player_of_Match"].mode()
```

```
ipl["Player_of_Match"].fillna("AB de Villiers", inplace= True)
```

```
#Mean value for margin column as it is a numerical column
ipl["Margin"].mean()
```

```
ipl["Margin"].fillna(17, inplace= True)
```

```
ipl.isnull().sum()
```

Missing values are cleaned

###Checking duplicates

```
ipl.shape
```

```
#Dropping duplicates
ipl.drop_duplicates(keep="first",inplace=True)
#checing shape again
ipl.shape
```

No duplicates values available in dataset and also now it has no missing values

###Handling outlier

```
q1=ipl["Margin"].quantile(0.25)
q2=ipl["Margin"].quantile(0.50)
q3=ipl["Margin"].quantile(0.75)
print("q1 is",q1)
print("q2 is",q2)
print("q3 is",q3)
```

```
#Calculating iqr
iqr=q3-q1
iqr
```

```
#Calculating minimum whisker
minimum_whisker=q1-1.5*iqr
minimum_whisker
```

```
maximum_whisker=q3+1.5*iqr
maximum_whisker
```

```
ipl.head()
```

```
len(ipl[ipl["Margin"]>36])  # So total in 136 columns there are outliers lets remove these
```

```
#Lets say maximum values is 40 beyond 40 will replace the values
#lets find the median
median_margine=ipl["Margin"].median()
median_margine
```

```
#Changinf extreme values with median
ipl.loc[ipl["Margin"] > 40,"Margin"]=median_margine
```

```
plt.figure(figsize=(10,6))
sns.boxplot(x=ipl["Margin"],color="gold") #Plotting box plot again
```

```
plt.figure(figsize=(10,8))
sns.displot(x=ipl["Margin"],kind="kde") #plotting desnsity plot again
```

So maxium value which should have this column is 36 let check the lenght beyond 36

Conclusion
1. Margin column has mean vlaue 17 with variance 450, means it has extreme values.
2. It has extreme values we can see in boxplot.But imputed with median value of tjje column.
3. margine is skewed in nature.
